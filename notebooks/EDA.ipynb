Univariate Anaysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['temperature', 'humidity', 'vibration_level', 'stock_quantity', 'supplier_rating', 'delivery_delay_days']
for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
```

```python
cat_cols = ['inventory_status', 'shipment_status', 'weather_condition', 'logistics_partner']
for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar', title=f'{col} Frequency')
```

```python
for col in num_cols:
    plt.figure()
    sns.boxplot(x='manual_risk_label', y=col, data=df)
    plt.title(f'{col} vs manual_risk_label')
```

```python
for col in cat_cols:
    plt.figure()
    sns.countplot(x=col, hue='manual_risk_label', data=df)
    plt.title(f'{col} distribution by Risk Level')
    plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```
